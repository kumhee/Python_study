```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False
```

```python
member_IDs = ['m_01', 'm_02', 'm_03', 'm_04']
before_ex = [27, 35, 40, 33]
after_ex = [30, 38, 42, 37]
```

```python
n_data = len(member_IDs)
index = np.arange(n_data)
plt.bar(index, before_ex, tick_label = member_IDs)
plt.show()
```

```python
colors=['r','g','b','m']
plt.bar(index, before_ex, tick_label = member_IDs, color=colors)
plt.show()
```

```python
colors=['r','g','b','m']
plt.bar(index, before_ex, tick_label = member_IDs, color=colors, width=0.1)
plt.show()
```

```python
colors=['r','g','b','m']
plt.barh(index, before_ex, tick_label = member_IDs, color=colors) # barh 뒤집혀서 출력
plt.show()
```

```python
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
# print(list(calender.month_name))

data = [
    [50, 40, 30, 20, 60, 70, 80, 90, 100, 110, 120, 130],
    [60, 50, 40, 30, 80, 90, 100, 110, 120, 130, 140, 150]
]

n_data = len(month)
index = np.arange(n_data)
before_ex = data[0]
after_ex = data[1]

bwidth = 0.3

plt.bar(index, before_ex, tick_label = month, align='edge', width=bwidth, label='2022년')
plt.bar(index + bwidth, after_ex, tick_label=month, color='c', align='edge', width=bwidth, label='2023년')
plt.xticks(index + bwidth, month)
plt.legend()
plt.xlabel('월')
plt.ylabel('판매량')
plt.title('서초구 상점 물건 판매량')
plt.show()
```
