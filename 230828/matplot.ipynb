```python
x = np.arange(-5, 5, 0.1)
y1 = x**2-2
y2 = 20*np.cos(x)**2

plt.figure(1)
plt.plot(x, y1)

plt.figure(2)
plt.plot(x, y2)

plt.figure(1)
plt.plot(x, y2)

plt.figure(2)
plt.clf()  # 기존에 있던 그래프 삭제 -> .clf()
plt.plot(x, y1)
```

```python
x = np.arange(0, 10 ,0.1)
y1 = 0.3*(x-5)**2 + 1
y2 = -1.5*x + 3
y3 = np.sin(x)**2
y4 = 10*np.exp(-x) + 1
```

```python
plt.subplot(2,2,1) # 한 페이지에 4개의 그래프가 전부 보이게 출력 -> .subplot()
plt.plot(x, y1)

plt.subplot(2,2,2)
plt.plot(x, y2)

plt.subplot(2,2,3)
plt.plot(x, y3)

plt.subplot(2,2,4)
plt.plot(x, y4)
```

```python
x = np.linspace(-4, 4, 100)
y1 = x**3
y2 = 10*x**2-2

plt.plot(x,y1,x,y2)
plt.xlim(-5, 5) # 범위지정 -> .xlim() .ylim()
plt.ylim(-100, 100)
plt.show()
```

```python
# 그래프 꾸미기
x = np.arange(0, 5, 1)
y1 = x
y2 = x+1
y3 = x+2
y4 = x+3
```

```python
plt.plot(x, y1, 'm', x, y2, 'y', x,y3, 'k', x,y4, 'b') # 색깔 지정해서 추가
plt.show() 
```

```python
plt.plot(x, y1, '-', x, y2, '--', x,y3, ':', x,y4, '-.') # 선 스타일 변경
plt.show()
```

```python
plt.plot(x, y1, 'o', x, y2, '^', x,y3, 's', x,y4, 'd') # 선 스타일 도형으로 변경 (점 그래프)
plt.show()
```

```python
plt.plot(x, y1, '>--r', x, y2, 's-g', x,y3, 'd:b', x,y4, '-.Xc') # >--r :우측방향 실선으로 red / ...
plt.show()
```

```python
x = np.arange(-4.5, 5, 0.5)
y = 2*x**3

plt.plot(x,y)
plt.title('TITLE')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid() # 표 상에서 그래프를 그려서 보여줌. -> .grid()
```

```python
x = np.arange(0, 5, 1)
y1 = x
y2 = x+1
y3 = x+2
y4 = x+3

plt.plot(x, y1, '>--r', x, y2, 's-g', x,y3, 'd:b', x,y4, '-.Xc')
plt.legend((['data1', 'data2', 'data3', 'data4']))  #data1-data4가 어떤 선인지 어떤스타일인지 따로 보여줌 
plt.show()
```

```python
plt.plot(x, y1, '>--r', x, y2, 's-g', x,y3, 'd:b', x,y4, '-.Xc')
plt.legend((['data1', 'data2', 'data3', 'data4']), loc='lower right') #.loc='lower right 오른쪽 아래쪽으로 지정.
plt.show()
```

```python
plt.plot(x, y1, '>--r', x, y2, 's-g', x,y3, 'd:b', x,y4, '-.Xc')
plt.legend((['data1', 'data2', 'data3', 'data4']), loc= 4)
plt.title('TITLE')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
```

```python
x = np.arange(0, 5, 1)
y1 = x
y2 = x+1

plt.plot(x,y1, 'r', x,y2, '--b')
plt.legend((['선 그래프1', '선 그래프2']))
plt.text(2.5, 7, "문자열 출력")

#축 라벨 설정
plt.xlabel('X축')
plt.ylabel('Y축')

#제목설정
plt.title('선 그래프 예시')

#눈금설정
plt.xticks([1,2,3,4])
plt.yticks([0,5,10])

#격자표시
plt.grid(True)

#차트 꾸미기
ax = plt.gca() # gca = get current axes : 현재 활성화된 축 객체을 가져온다.
ax.spines["top"].set_visible(False) # - 스파인
ax.spines["right"].set_visible(False) # 위, 오른쪽은 삭제.
ax.spines["bottom"].set_linewidth(3) # 왼쪽, 아래는 선 두께를 3으로 설정.
ax.spines["left"].set_linewidth(3)
ax.tick_params(width = 3) # params 축 굵기는 3으로 설정.

#차트 출력
plt.show()
```

```python
city = ['서울', '인천', '대전', '대구', '울산', '부산', '광주']

lat = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16]
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.09, 126.85]

pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]

size = np.array(pop_den)
color = ['r', 'g', 'b', 'c', 'm', 'k', 'y']

plt.scatter(lon, lat, s=size, c=color)
plt.xlabel('경도(longitude)')
plt.ylabel('위도(latitude)')
plt.title('지역별 인구밀도')

for x,y, name in zip(lon, lat, city):
    plt.text(x,y,name)
    
plt.show
```
